#### Machine Learning Classification 
1. Classification 개요 
    - Supervised Learning는 Label 즉 명시적인 정답이 있는 데이터가 주어진 상태에서 학습하는 Machine-Learning 방식 
    - Supervised Learning의 대표적인 유형인 분류(Classifictation)는 학습 데이터로 주이진 데이터의 Feature, Label 값을 학습해 학습 모델 생성 
    - 생성된 학습 모델을 기준으로 새로운 데이터값이 주어졌을 때 미지의 Label값을 예측하는 것 
    - Classification은 다양한 Machine Learning Algorithm으로 구현이 가능 
        
        1. Naive Bayes   
        2. Logistic Regression  
        3. Decision Tree 
        4. Support Vertor Machine 
        5. Nearest Neighbor 
        6. Neural Network 
        7. Ensemble


2. Decision Tree 
    -  ML 알고리즘 중 직관적으로 이해하기 쉬운 알고리즘
    - 데이터에 있는 규칙을 학습을 통해 찾아내 Tree 기반의 분류 규칙을 만듦
    - 일반적으로 규칙을 가장 쉽게 표현하는 방법은 IF/Else 기반으로 나타내는 것
    - 데이터의 어떠한 기준을 바탕으로 규칙을 만들어야 가장 효율적인 분류가 될 것인가가 알고리즘의 성능을 크게 좌우한다.

#### 1. Import Library 

```python
import numpy as np 
import pandas as pd 
import sklearn 
import seaborn as sns 
import matplotlib.pyplot as plt 
import graphviz 

from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, recall_score, precision_score
from sklearn.datasets import load_iris 
from sklearn.tree import DecisionTreeClassifier, export_graphviz 
```

##### 2.Load dataset 


```python
iris_data = load_iris()
print(type(iris_data))
```

#### 3.train test split 

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=121)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### 4.Decision Tree Classifiter 


```python
dt_clf = DecisionTreeClassifier(random_state=121)
dt_clf.fit(X_train, y_train)
# print(type(dt_clf))
```

```python
export_graphviz(dt_clf, out_file="E:\\result\\tree.dot", class_names=iris_data.target_names, feature_names=iris_data.feature_names, impurity=True, filled=True)
```

```python
with open("tree.dot") as tree_files :
    dot_praph = tree_files.read()

graphviz.Source(dot_praph)
```
