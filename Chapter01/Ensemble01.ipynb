#### Ensemble 
- Voting Classifier 
- Dataset : Wisconsin Diagnostoc Breast Cancer Dataset 



```python
import pandas as pd 
import numpy as np 

from sklearn.ensemble import VotingClassifier 
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
```

```python
cancer_data = load_breast_cancer()
```

```python
df_data = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
```

```python
df_data.info()
```

```python
df_data.head(10)
```

#### Create Model 
- LogisticRegression 
- KNeighborsClassifier 


```python
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=8)
```

```python
vo_clf = VotingClassifier(estimators=[("LR", lr_clf), ("KNN", knn_clf)], voting="soft")
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, test_size=0.2, random_state=121)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
vo_clf.fit(X_train, y_train)
predict_result = vo_clf.predict(X_test)
print("Accuracy of Voting Classifier {0:.4f}".format(accuracy_score(y_test, predict_result)))
```

```python
classifiers = [lr_clf, knn_clf]

for classifier in classifiers:
    classifier.fit(X_train, y_train)
    predict_result = classifier.predict(X_test)
    class_name = classifier.__class__.__name__
    print("{0} Accuracy {1:.4f}".format(class_name, accuracy_score(y_test, predict_result)))
```
